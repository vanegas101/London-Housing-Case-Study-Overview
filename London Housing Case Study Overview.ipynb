```python
def create_price_ratio(df, borough):
    price_1998 = df.loc[borough]['1998-01-01']
    price_2018 = df.loc[borough]['2018-01-01']
    ratio = price_1998 / price_2018
    return ratio

boroughs = test2.index.unique()
for borough in boroughs:
    ratio = create_price_ratio(test2, borough)
    print(f"{borough}: {ratio}")
```
